```python
from transformers import AutoTokenizer,AutoModelForTokenClassification
from transformers import pipeline
```

```python
tokenizer = AutoTokenizer.from_pretrained("dslim/bert-base-NFR")
model = AutoModelForTokenClassification.from_pretrained("dslim/bert-base-NFR")
nlp = pipeline("ner", model=model, tokenizer=tokenizer)
text ="As the sun dipped below the horizon, casting long shadows across the golden fields, a lone traveler with a weathered backpack and worn boots approached the quaint village nestled amidst the rolling hills, drawn by the promise of a warm meal and a friendly face after a long day on the dusty road."
```

```python
ner_results = nlp(text)
for ne in ner_results: 
    print(ne)
```
