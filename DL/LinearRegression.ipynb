```python
import numpy as np
import matplotlib.pyplot as plt
N =100
X = np.random.rand(N)
Y = 5 * X + 10 + 0.4 * np.random.rand(N)
plt.plot(X, Y, '.')
plt.show()
```

```python
import tensorflow as tf

W = np.random.rand()
b = np.random.rand()

W = tf.Variable(W)
b = tf.Variable(b)

for epoch in range(1000):
    with tf.GradientTape() as tape:
        y = W * X + b
        loss = tf.reduce_mean(tf.square(y - Y))
    grads = tape.gradient(loss, [W, b])
    W.assign_sub(grads[0] * 0.01)
    b.assign_sub(grads[1] * 0.01)
    print(epoch,W.numpy(),b.numpy(),loss.numpy())
```

```python
Z = W * X + b
plt.plot(X, y, 'b')
plt.plot(X, Z, 'r')
plt.show()
```
