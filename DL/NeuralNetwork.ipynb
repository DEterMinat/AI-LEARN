```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def perceptron(X, T, lr=0.1, w=None):
    X = np.hstack((np.array(X), np.ones(len(X), dtype=int)[:, np.newaxis]))
    if w is None:
        w = X[np.random.randint(0,len(X))].copy()
    done = False
    while not done:
        done = True
        for i, x in enumerate(X):
            if T[i]*np.dot(x, w) <= 0:
                w+= lr * x * T[i]
                done = False
    return w
```

```python
def plot_hyperplane2d(X, T, w):
    X = np.array(X)
    T = np.array(T)
    plt.plot(X[T==1,0], X[T==1,1], 'go')
    plt.plot(X[T==-1,0], X[T==-1,1], 'ro')
    xlim = plt.gca().get_xlim()
    slope = -w[0] / w[1]
    bias = -w[-1] / w[1]
    plt.plot(xlim, [xlim[0]*slope + bias, xlim[1] * slope + bias ], 'b')
```

```python
X = [[1,1], [1,-1], [-1,-1], [-1,-1]]
T = [1,1,1,1]
```

```python
w = perceptron(X , T , lr=1e-1)
print(w)
plot_hyperplane2d(X, T, w)
```
