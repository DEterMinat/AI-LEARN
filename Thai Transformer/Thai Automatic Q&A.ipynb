```python
from datasets import load_dataset
thaiqa_ds = load_dataset('thaiqa_squad')
```

```python
import  pandas as pd
qa_train = pd.DataFrame(thaiqa_ds["train"])
```

```python
from lxml import etree

parser = etree.XMLParser(recover=True)
tree = etree.fromstring(qa_train["context"][0], parser=parser)
print(tree.tag)
print(tree.attrib)
```

```python
def extract_text(body):
    tree = etree.fromstring(body, parser=parser)  # Use etree.fromstring
    return tree.text
```

```python
extract_text(qa_train["context"][0])
```

```python
qa_train["context_text"] = qa_train["context"].apply(extract_text)
```

```python
qa_train.head()
```

```python
from pythainlp.tokenize import word_tokenize

qa_train["context_text_wordsegged"] = qa_train["context_text"].apply(
    lambda x: " ".join(word_tokenize(x))  # Apply word_tokenize first, then join
)
```

```python
qa_train["context_text_wordsegged"] = qa_train["question"].apply(lambda x: " ".join(word_tokenize(x)))
```

```python
qa_train.head()
```

```python
qa_train["answers_wordsegged"] = qa_train["text"].apply(
    lambda x: " ".join(word_tokenize(x))  # Use "text" if answers are there
)
```

```python
qa_train.head()
```

```python
answer_starts = []
for idx,row in qa_train.iterrows():
    try:
        answer_starts = row["context_text_wordsegged"].index(row["answers_wordsegged"])
    except:
        answer_starts = -1
    answer_starts.append(answer_starts)
```

```python
qa_train[["answer_starts"]] = answer_starts
```
